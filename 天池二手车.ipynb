```python
X_train['notRepairedDamage']=X_train['notRepairedDamage'].astype('float')
```

```python
from sklearn.model_selection import KFold, GridSearchCV
from lightgbm.sklearn import LGBMRegressor
from sklearn.metrics import mean_absolute_error

```

```python
clf = LGBMRegressor(n_estimators=10000, learning_rate=0.05,bagging_fraction= 0.7, feature_fraction= 0.9, max_depth=15)

kf = KFold(n_splits=5, shuffle=True, random_state=2020)

maes = []
for i, (trn_idx, val_idx) in enumerate(kf.split(X_train, Y_train)):
    print('--------------------- {} fold ---------------------'.format(i+1))
    trn_x, trn_y = X_train.iloc[trn_idx].reset_index(drop=True), Y_train[trn_idx]
    val_x, val_y = X_train.iloc[val_idx].reset_index(drop=True), Y_train[val_idx]
    clf.fit(trn_x, trn_y)
    pred_y = clf.predict(val_x)
    mae = mean_absolute_error(np.expm1(val_y), np.expm1(pred_y))
    maes.append(mae)
    print('mae: ', mae)
print('average mae: ', np.mean(maes))
```

```python

X_test['notRepairedDamage']=X_test['notRepairedDamage'].astype('float')
sub = test[['SaleID']].copy()
sub['price'] = np.expm1(clf.predict(X_test)) 
sub.to_csv('sumbitusedcar.csv',index=False)
```
